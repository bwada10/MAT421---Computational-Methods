**1.2: Eigenvalues And Eigenvectors**

Any λ that satisfies, Ax =λx, is the **Eigenvalue** of the matrix A and x is the **Eigenvector** corresponding to λ where matrix A is n x n matrix. Eigenvalues and eigenvectors are widely used in many areas of mathematics, science, and engineering.

Python allows us to determine eigenvalues and eigenvectors with ease.

---



> *Find the Eigenvalue and Eigenvector for matrix A:*


> Capture2.PNG





```python
import numpy as np
from numpy.linalg import eig

MatrixA = np.array([[2, 2, 4], 
                  [1, 3, 5],
                  [2, 3, 4]])
w,v=eig(MatrixA)
print('Eigenvalue:', w)
print('Eigenvector:', v)
```

**1.3: Least Squares Regression**

**Least Squares Regression** is a statistical method for fitting a line to a set of data points. The line best approximates the data by *minimizing* the sum of squared differences between the observed data points and the values predicted by the line.

---



> *Do a least square regression with an estimation function defined by y = a_1*x + a_2, given the created artificial data:*





```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = np.linspace(0, 1, 101)  #Artificial data created
y = 1 + x + x * np.random.random(len(x)) #Artificial data created

A = np.vstack([x, np.ones(len(x))]).T #Matrix A

# turn y into a column vector
y = y[:, np.newaxis]
```

```python
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)  #Least square regression
print(alpha)
```

```python
plt.figure(figsize = (10,8))  #Plot the data points and least square regression
plt.plot(x, y, 'b.')  
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
