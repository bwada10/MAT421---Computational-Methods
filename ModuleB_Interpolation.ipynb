**Chapter 17 - Interpolation**

***17.1: Interpolation Problem Statement***

Given many reliable data points consisting independent (x_i) and dependent (y_i) data values, where i = 1,...,n, we want an estimation function that goes through the data points called an **Interpolation Function**.


***17.2: Linear Interpolation***

For **Linear Interpolation**, the estimated data points are to be *assumed* to lie on the line of the nearest data points. 

WLOG, assume that the *x*-data points are in ascending order, x_i < x_i+1, and x be a point between x_i < x < x_i+1. Hence the Linear Interpolation formula at x is: y(x) = y_i + [(y_i+1 - y_i)(x - x_i)]/(x_i+1 - x_i).


---

> *Find the Linear Interpolation at x = 5.5 based on the data x = [1,2,4], y =[2,9,6]:*

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [1, 2, 4] #X-Data Points
y = [2, 9, 6] #Y-Data Points

f = interp1d(x, y)  #Make a Interpolation function y in terms of x.
y_hat = f(3.5)
print(y_hat)
```

```python
plt.figure(figsize = (10,8))  #Plot Interpolation
plt.plot(x, y, '-ob')
plt.plot(3.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 3.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

***17.3: Cubic Spline Interpolation***

A **Cubic Spline Interpolation**, is a set of piecewise cubic functions. To find the interpolating function, we must first determine the coefficients a_i,b_i,c_i,d_i for each of the cubic functions. For **n** points, there are **n−1** cubic functions to find, and each cubic function requires **four** coefficients. Therefore we have a total of **4(n−1)** unknowns, and so we need **4(n−1)** independent equations to find all the coefficients.


---
> *Find the cubic interpolation of the data set x= [1,2,3], y = [2,9,6] for 0 < x < 2:*

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [0, 1, 2] #X-Data Points
y = [2, 9, 6] #Y-Data Points

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
