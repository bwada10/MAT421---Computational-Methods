**22.1: ODE Initial Value Problem Statement**

An **Ordinary Differential Equation** is a differential equation where the independent variable, also the derivatives, is in one dimension.

In contrast, a **Partial Differential Equation** is a general form differential equation where $x$ is a vector containing the independent variables $x_1,x_2,x_3,...,x_m$, and the partial derivatives can be of any order and with respect to any combination of variables.


**22.2: Reduction of Order**

When solving higher order differential equations, often we must **reduce the order** of the differential equation to first order. By reducing the order, we can often simplify the equation and make it easier to work with. In many cases, it may be easier or more efficient to solve a lower order differential equation than a higher order one.

**22.3: The Euler Method**

The **Euler Method** approximates the solution of an ODE by using the derivative of the function at a particular point to estimate the value of the function at a nearby point. The idea is to take small steps in the direction of the derivative of the function, using the initial value as a starting point. At each step, the derivative is evaluated at the current point, and the solution is updated by adding the product of the derivative and a small step size. 

Forumual for the Euler Method:

$y_{n+1}=y_n+hf(x_n,y_n)$
---


> *The differential equation $\frac{df(t)}{dt}=e^{−t}$ with initial condition $f0=−1$ has the exact solution $f(t)=−e^{−t}$. Approximate the solution to this initial value problem between 0 and 1 in increments of 0.1 using the Explicit Euler Formula. Plot the difference between the approximated solution and the exact solution.*



```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline


f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
